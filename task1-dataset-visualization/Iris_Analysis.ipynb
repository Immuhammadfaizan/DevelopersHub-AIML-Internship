```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('./data/iris.csv')
df.head()
```

```python
print("Dataset Shape:", df.shape)
print("\nDataset Column: ", df.columns)
print("\nDataset First Few Rows: \n", df.head())
```

```python
# Information
print("\n\nINFO")
print("=" * 50)
print(df.info())
print("=" * 50)

# Summary Statistics
print("\n\nSTATISTICS")
print("=" * 50)
print(df.describe())
print("=" * 50)

# Checking Missing values
print("\n\nCHECKING MISSING VALUES")
print("=" * 50)
print(df.isnull().sum())
print("=" * 50)
```

```python
# Visualize Dataset, Scatter Plot for Petal Length vs Petal Width using Seaborn

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='petal length (cm)',
    y='petal width (cm)',
    hue='species'
)

plt.title("Petal Length vs Petal Width")
plt.savefig("./images/petal_length_vs_petal_width.png", dpi=150)
plt.show()
```

```python
# pair plot for all the features

sns.pairplot(
    df,
    hue='species'
) 

plt.savefig("./images/pair_plot.png", dpi=150)
plt.show()
```

```python
# Histogram

df.hist(
    figsize=(10,8), # to set the size of the plot (10=width, 8=height)
    bins=15 # to set the number of bins (the number of bars in the histogram) 
)

plt.suptitle("Histogram of Features", fontsize=16)
plt.tight_layout() # to avoid overlapping
plt.savefig("./images/histogram_of_features.png", dpi=150)
plt.show() 
```

```python
# seaborn histogram (used for sepal length distribution)

plt.figure(figsize=(8,6))

sns.histplot(
    data=df,
    x='sepal length (cm)',
    kde=True
)

plt.title("Sepal Length Distribution")
plt.savefig("./images/seaborn_histogram_sepal_length.png", dpi=150)
plt.show()
```

```python
# Identify the outliers using box plot for sepal length

plt.figure(figsize=(8,6))

sns.boxplot(data=df)

plt.title("Box Plot of Features")
plt.savefig("./images/outliers_using_boxing_plots.png", dpi=150)
plt.show()
```

```python
# feature wise box plot for petal length by species

plt.figure(figsize=(8,6))

sns.boxplot(
    x='species',
    y='petal length (cm)',
    data=df
)

plt.title("Petal Length by Species")
plt.savefig("./images/petal_length_box_plot.png", dpi=150)
plt.show()
```

```python
# correlation matrix 

numeric_df = df.drop('species', axis=1)

corr = numeric_df.corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Matrix")
plt.savefig("./images/correlation_matrix.png", dpi=150)
plt.show()
```

## Observations

1. Iris dataset contains 150 samples and 5 columns.
2. No missing values were found.
3. Petal length and petal width are highly correlated.
4. Setosa species is clearly separated from other species.
5. Histograms show approximately normal distributions.
6. Box plots reveal a few potential outliers in sepal measurements.
7. Petal measurements provide better species separation than sepal measurements.
